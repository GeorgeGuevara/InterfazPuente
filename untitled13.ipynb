```python
import socket
import threading
import time
import random

# Configuración del servidor
HOST = 'localhost'  # Dirección IP del servidor
PORT_START = 8889  # Puerto inicial para intentar
NUM_CLIENTS = 10  # Número de clientes ficticios a simular

# Función para manejar la interacción con un cliente de automóvil
def handle_client(conn, addr):
    print(f"Conexión establecida con {addr}")

    while True:
        try:
            # Recibir comando del cliente
            data = conn.recv(1024).decode('utf-8')
            if not data:
                break

            print(f"Comando recibido de {addr}: {data}")

            # Simular cruce del puente (lógica a implementar)
            # ...

            # Enviar actualización al cliente (lógica a implementar)
            # ...

        except Exception as e:
            print(f"Error en la conexión con {addr}: {e}")
            break

    conn.close()
    print(f"Conexión cerrada con {addr}")

# Función para encontrar un puerto disponible
def find_available_port(start_port):
    port = start_port
    while True:
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
            try:
                s.bind((HOST, port))
                return port
            except OSError as e:
                if e.errno == 98:  # Address already in use
                    port += 1
                else:
                    raise

# Función para simular un cliente ficticio
def simulate_client(client_id):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        try:
            s.connect((HOST, PORT))
            print(f"Cliente ficticio {client_id+1} conectado.")
            time.sleep(random.randint(1, 10))  # Simula un tiempo de conexión aleatorio
            s.close()
            print(f"Cliente ficticio {client_id} desconectado.")
        except Exception as e:
            print(f"Error al conectar el cliente ficticio {client_id}: {e}")

# Función para iniciar el servidor
def start_server():
    global PORT  # Declaramos PORT como global para poder modificarlo dentro de la función
    PORT = find_available_port(PORT_START)  # Encontrar un puerto disponible
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        s.bind((HOST, PORT))
        s.listen()
        print(f"Servidor escuchando en {HOST}:{PORT}")

        while True:
            conn, addr = s.accept()
            client_thread = threading.Thread(target=handle_client, args=(conn, addr))
            client_thread.start()

# Inicia el servidor en un hilo separado
server_thread = threading.Thread(target=start_server)
server_thread.start()

# Espera un momento para que el servidor se inicie
time.sleep(1)

# Crea múltiples hilos para simular clientes concurrentes
threads = []
for i in range(NUM_CLIENTS):
    thread = threading.Thread(target=simulate_client, args=(i,))
    threads.append(thread)
    thread.start()

# Espera a que todos los hilos terminen
for thread in threads:
    thread.join()

print("Todos los clientes han terminado.")
```
